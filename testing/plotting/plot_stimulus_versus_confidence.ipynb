```python
cfg = remeta.Configuration()
cfg.param_type2_criteria.enable = 3
params = dict(
    type1_noise=0.5,
    type1_bias=0.2,
    type2_noise=0.2,
    type2_criteria=[0.25, 0.5, 0.75]
)
ds2 = remeta.simulate(nsubjects=1, nsamples=1000, params=params, squeeze=True, stim_levels=4, cfg=cfg, silence_warnings=True)
```

```python
remeta.plot_stimulus_versus_confidence(ds2)
```

```python
remeta.plot_stimulus_versus_confidence(ds2.stimuli, ds2.confidence, probability_correct=True, params=ds2.params, model_prediction=True)
```

```python
remeta.plot_stimulus_versus_confidence(ds2.stimuli, ds2.confidence, probability_correct=True, params=ds2.params, model_prediction=True, model_prediction_disable_type2_noise=True)
```
